But CNN architectures do have some design principles:

*Controlling the receptive field*

For a given neuron in a hidden layer $l$:

- spatial area (patch) in original image that this neuron has access to in computation
    

Want detector to incorporate large patch in making a decision:



- VGG-16 and VGG-19 are CNN-based models designed with aforementioned principles in mind
- first modern, standardized CNNs
- clear demonstration of the power of *depth* in a neural network

#### Transfer Learning

- VGG16's CNN block can be thought of as feature extractor
- weights trained: extract useful feature maps for classification on ImageNet

Are the VGG filters learned on Imagenet possibly useful as feature extractors on different data sets?

**Let's take a look for ourselves:**

```python
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt
import numpy as np
```

Function takes in image and returns feature maps computed from first CNN layer of VGG16:

```python
def image_layerfeaturedisplayer(vgg_mod, image_path, pixel_width, layer_num, feature_idx = 0):
    
    image_obj = load_img(image_path, target_size = (pixel_width,pixel_width)) # maps input image to size requested by VGG
    image_array = img_to_array(image_obj)
    
    # pass into VGG preprocessor (scales, centers, etc, REQUIRED)
    processed_img = preprocess_input(image_array)[np.newaxis, :]
    
    
    # the .output attribute returns the feature map (Tensor) returned by CNN layer
    firstlayer_mapper = Model(inputs = vgg_mod.inputs, outputs = vgg_mod.layers[layer_num].output)
     
    feature_maps = tf.squeeze(firstlayer_mapper(processed_img))
    
    # num_acivations
    num_features = feature_maps.shape[-1]
    
    fmap = feature_maps[:,:,feature_idx] 
    kernel_weights = vgg_mod.layers[layer_num].get_weights()[0][:,:,:,feature_idx]

    fig = plt.figure(layout="constrained", figsize = (10,5))
    subfigs = fig.subfigures(1, 3, wspace=0.07, width_ratios=[1.5, 1., 1.5])
    axs0 = subfigs[0].subplots()
    subfigs[0].suptitle('Original Image')
    axs0.imshow(image_obj)

    axs1 = subfigs[1].subplots(3, 1)
    subfigs[1].suptitle('Feature Kernel')
    axs1[0].imshow(kernel_weights[:,:,0])
    axs1[1].imshow(kernel_weights[:,:,1])
    axs1[2].imshow(kernel_weights[:,:,2])
    
    axs2 = subfigs[2].subplots()
    subfigs[2].suptitle('Feature Map')
    axs2.imshow(fmap)
```

```python
image_layerfeaturedisplayer(vgg_model_full,
                    'images/me.jpg', 224, 1, 50)
```

```python
image_layerfeaturedisplayer(vgg_model_full,
                    'images/me.jpg', 224, 1, 9)
```

```python
image_layerfeaturedisplayer(vgg_model_full,
                    'images/me.jpg', 224, 1, 60) 
```

VGG16 pretrained early layers are low level feature detectors:
- capturing useful geometric shapes, boundaries, textures in image

- note: dimensions of high level feature map much smaller than original image

**VGG-16, CONV-part of network, trained on large dataset:**
- **contains a large quantity (high capacity network) of generally useful feature constructors for images**

**Transfer Learning**

*Leverage what has been learned*
- Take only Conv-layers of VGG and use as pre-trained feature extractor
- Freeze weights (make them untrainable)

*Out of the box: many feature detectors learned on large, general image dataset useful for more specific tasks*

<center><img src = "images/transfer_learning_ships.png"></center>

**Steps**
- Need to chop off FCN part of VGG 
- Freeze conv-layers
- Add our own layers
